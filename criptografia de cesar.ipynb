```python
tipo = input("Selecione qual modo você deseja, entre encripitar (1) e desencriptar (2): ")
numero_de_rotacoes = int(input("Digite o numero de rotações: "))
alfabeto = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']


if tipo == "1":
    texto1 =  input("Insira o texto para encriptar: ")
    texto = texto1.lower()
    palavra = ''
    for c in texto:
      if c in alfabeto:
        c_index = alfabeto.index(c) #
        palavra += alfabeto[(c_index + numero_de_rotacoes) %len(alfabeto)] 
      else:                                                                 
        palavra += c                                                       
    print("A palavra sairá como: ",palavra)

elif tipo == "2":
    texto1   =  input("Insira o texto para descriptografar: ")
    texto = texto1.lower()
    palavra = ''
    for c in texto:
      if c in alfabeto:
        c_index = alfabeto.index(c)
        palavra += alfabeto[(c_index - numero_de_rotacoes) %len(alfabeto)] 
      else:
        palavra += c
    print("A palavra sairá como: ",palavra)

else:
  print("Você digitou uma opção inválida")

```
